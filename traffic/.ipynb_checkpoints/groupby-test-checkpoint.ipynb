```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
sns.set(rc={'figure.figsize':(18,12)})
```

```python
df = pd.read_csv('mgido-calls-2017-2020.csv')
df.columns = ['date', 'income', 'subst', 'inner', 'wait', 'length']
# DRY :)
for field in ('income', 'subst', 'inner'):
    df[field] = df.apply({field: lambda x: x[1:] if x.startswith('+') else x})
df.date = pd.to_datetime(df.date)    
df.head()
```

```python
df.date = pd.to_datetime(df.date)
df.head()
```

```python
df_wide = df.join(pd.get_dummies(df['subst']))
```

```python
df_res = df_wide.set_index('date').resample('1d').sum()
```

```python
df_plot = df_res.rolling(20, min_periods=1).sum()
```

```python
sns.lineplot(data=df_plot.iloc[:,:], palette="tab10", linewidth=1)
```
